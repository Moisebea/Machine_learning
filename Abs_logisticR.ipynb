```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Absenteeism_preprocessed.csv')
data.head()
```

```python
targets = np.where(data['Absenteeism Time in Hours'] > 
                   data['Absenteeism Time in Hours'].median(), 1, 0)
```

```python
targets
```

```python
data['Excessive Absenteeism'] = targets
data.head(10)
```

```python
targets.sum() / targets.shape[0]
```

```python
data_with_target = data.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data_with_target.head(10)
```

```python
data_with_target.shape
```

```python
data_with_target.iloc[:,:-1]
```

```python
unscaled_inputs = data_with_target.iloc[:,:-1]
```

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train , x_test , y_train , y_test = train_test_split(scaled_inputs,targets,train_size=.8 , random_state=20)
```

```python
print(x_train.shape ,y_train.shape)
```

```python
print(x_test.shape ,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
model_outputs = reg.predict(x_train)
model_outputs
```

```python
y_train
```

```python
model_outputs == y_train
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature name'],data = feature_name)

summary_table['Coefficient'] =np.transpose(reg.coef_)

summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```
